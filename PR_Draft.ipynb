```python
import pandas as pd

import glob

```

```python
#merging all the cdc_pram data files to get one data frame

path = r'C:\Users\pakan\PycharmProjects\PRFinalProject\files' 

cdc_pram_files = glob.glob(path + "/CDC_PRAMStat_Data_for_*.csv")

li = []

for filename in cdc_pram_files:

    df = pd.read_csv(filename, index_col=None, header=0)

    li.append(df)

cdc_pram_frame = pd.concat(li, axis=0, ignore_index=True)
cdc_pram_frame
```
